Project overview: Classification of breast cancer diagnoses

Aim of the project: The aim of this data science project is to develop a predictive model that can differentiate between malignant and benign breast tumors based on digitized image features of breast masses. The features originate from fine needle aspirates and include dimensions such as radius, texture and circumference of the cell nuclei.

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

data_path = '/Users/caner/Desktop/Data Science/BreastCancerDiagnosis/data.csv'
data_df = pd.read_csv(data_path)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(6, 4))
sns.countplot(data=data_df, x='diagnosis', palette='coolwarm')
plt.title('(Distribution of diagnosis (M = malignant, B = benign)')
plt.show()
```

```python
sns.set(style="whitegrid")

selected_features = data_df.columns.drop(['id', 'diagnosis'])  

for feature in selected_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(data_df[feature], kde=True, color='skyblue')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
selected_features = ['radius_mean', 'concavity_mean', 'area_mean', 'smoothness_mean','concave points_mean',
                        'perimeter_se','concave points_worst','perimeter_worst','radius_worst','area_worst']

for feature in selected_features:
    plt.figure(figsize=(8, 4))
    sns.violinplot(x='diagnosis', y=feature, data=data_df, palette='coolwarm')
    plt.title(f'Violin plot of {feature} by diagnosis')
    plt.xlabel('Diagnosis')
    plt.ylabel(feature)
    plt.show()
```

```python
for feature in selected_features:
    plt.figure(figsize=(8, 4))
    sns.kdeplot(data_df[data_df['diagnosis'] == 'M'][feature], label='Malign', shade=True, color='red')
    sns.kdeplot(data_df[data_df['diagnosis'] == 'B'][feature], label='Benign', shade=True, color='blue')
    plt.title(f'Density plot of {feature} by diagnosis')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```
